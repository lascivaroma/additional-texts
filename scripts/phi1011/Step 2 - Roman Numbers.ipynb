```python
import regex

with open("./data/phi1011/phi001/phi1011.phi001.lascivaroma-lat1.xml") as f:
    text = f.read()
```

```python
NUMB = regex.compile("subtype=\"compositio\" n=\"(\w+)\"")

def roman_number(match):
    n = match.groups(1)[0]
    num = 0
    
    for char, value in [
        ("I", 1),
        ("H", 2),
        ("J", 1),
        ("V", 5),
        ("XL", 40),
        ("L", 50),
        ("X", 10),
        ("C", 100)
    ]:
        num += n.count(char) * value
        n = n.replace(char, "")
    return f"subtype=\"compositio\" n=\"{num}\""

with open("./data/phi1011/phi001/phi1011.phi001.lascivaroma-lat1.xml", "w") as f:
    f.write(NUMB.sub(roman_number, text))
```
