```python
import requests
import bs4

REPO = requests.get("http://theobi.com/Religion/Vulgate/")
data = bs4.BeautifulSoup(REPO
                         .text)
```

```python
from urllib.parse import urljoin

texts = {
    
}
for a in data.find_all('a', href=True):
    if a["href"].endswith(".txt") or a["href"].endswith(".vul"):
        texts[a["href"]] = requests.get(urljoin("http://theobi.com/Religion/Vulgate/", a["href"]))
```

```python
import regex
from collections import defaultdict

verse = regex.compile("^(\d+):(\d+)")
out = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: "")))

for filename, response in texts.items():
    #print(response.text)
    for line in response.text.split("\n"):
        if not line:
            continue
        if line[0].isnumeric():
            chap, ver = verse.findall(line)[0]
            line = verse.sub("", line)
        out[filename][chap][ver] += " " + line
```
